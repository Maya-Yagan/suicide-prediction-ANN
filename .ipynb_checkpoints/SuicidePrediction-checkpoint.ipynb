```python
import pandas as pd
import numpy as np

#load dataset
data = pd.read_csv('combined_processed_data.csv')

#display the first few rows
print("\n The first 5 rows:")
print(data.head())

#checking for empty cells
print("\n Null data number:")
print(data.isnull().sum())

#checking data types
print("\n Data types:")
print(data.info())

#viewing a summary of the dataset
print("\n Basic statistics:")
print(data.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlations
correlation_matrix = data.corr()

# Focus on the correlation with the target variable (Death rate per 100 000 population)
target_correlation = correlation_matrix['Death rate per 100 000 population'].sort_values(ascending=False)
print(target_correlation)

# Plot heatmap (darker red cells mean strong, lighter ones mean weak correlation)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
#since the age groups 0 and 1-4 don't contribute much to the dataset, they will be removed

# Drop columns '0 Year' and '1-4 Years'
data = data.drop(['0 Year', '1-4 Years'], axis=1)

# Verify the changes
print("\n Columns after dropping '0 Year' and '1-4 Years':")
print(data.columns)

correlation = data.corr()
print(correlation['Death rate per 100 000 population'])
```

```python
#after removing the 0 and 1-4 age groups, check the heatmap again

# Calculate correlations
correlation_matrix = data.corr()

# Focus on the correlation with the target variable (Death rate per 100 000 population)
target_correlation = correlation_matrix['Death rate per 100 000 population'].sort_values(ascending=False)

# Print correlations
print(target_correlation)

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix After Dropping 0 and 1-4 Years Age Groups')
plt.show()
```

```python
# features are the parameters that the prediction will be based on
features = data[['Sex', 'Country Name', '5-9 Years','10-14 Years','15-19 Years', '20-24 Years', '25-29 Years',
                 '30-34 Years', '35-39 Years', '40-44 Years', '45-49 Years', '50-54 Years', '55-59 Years', 
                 '60-64 Years', '65-69 Years', '70-74 Years', '75-79 Years', '80-84 Years', '85+ Years']]
#target is the output of the prediction
target = data['Death rate per 100 000 population']
```

```python
#Convert categorical data into numeric values
features = pd.get_dummies(features, columns=['Sex', 'Country Name'], drop_first=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# MinMaxScaler scales each feature to a specified range (default is 0 to 1)
scaler = MinMaxScaler()

# Fit the scaler to the input features and transform the features into a 0-1 range
# This ensures that all features have the same scale
features_scaled = scaler.fit_transform(features)

# - test_size=0.2 means that 20% of the data will be used as a test set, the reamining 80% will be for training
# - random_state=42 ensures reproducibility, so the data split remains consistent across different runs
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)
```

```python
#model 1
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout

# Define the model
model1 = Sequential()

# Input Layer
model1.add(Dense(32, input_dim=X_train.shape[1], activation='relu'))

# Hidden Layers
model1.add(Dense(16, activation='relu'))  # Hidden layer 1
# Dropout layer to prevent overfitting by randomly setting 20% of neurons to zero during training
model1.add(Dropout(0.2))
model1.add(Dense(8, activation='relu'))  # Hidden layer 2

# Output Layer
model1.add(Dense(1, activation='linear'))  # Regression task

# Compile the model
model1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='mean_squared_error',
              metrics=['mean_absolute_error'])

# Print the model summary
model1.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
# EarlyStopping helps stop training a model early if the performance on the validation set stops improving

# - monitor='val_loss': monitors the validation loss during training
# - patience=5: allows training to continue for 5 more epochs after the last improvement in validation loss
# - restore_best_weights=True: restores the model weights from the epoch with the best validation loss
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# - epochs=500: maximum number of epochs for training (stops earlier if EarlyStopping is triggered)
# - batch_size=32: number of samples processed before updating the model weights
# - validation_split=0.2: reserves 20% of the training data for validation 
# - callbacks=[early_stopping]: ends training if validation loss stops improving
history = model1.fit(X_train, y_train, epochs=500, batch_size=32,
                    validation_split=0.2, callbacks=[early_stopping])
```

```python
import matplotlib.pyplot as plt

# Plot Loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss Over Epochs: Model 1')
plt.show()

# Plot MAE
plt.plot(history.history['mean_absolute_error'], label='Training MAE')
plt.plot(history.history['val_mean_absolute_error'], label='Validation MAE')
plt.xlabel('Epochs')
plt.ylabel('Mean Absolute Error: Model 1')
plt.legend()
plt.title('MAE Over Epochs')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Predict on the test set
y_pred = model1.predict(X_test)

# Scatter plot of true vs predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.6, edgecolor='k')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--')  # Line y=x
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('True vs Predicted Suicide Rates: Scatter Plot for Model 1')
plt.show()

# Line Plot
plt.figure(figsize=(12, 6))
plt.plot(range(len(y_test)), y_test, color='red', linestyle='-', label='True Values')
plt.plot(range(len(y_pred)), y_pred, color='blue', linestyle='-', label='Predicted Values')
plt.title("True vs Predicted Values: Line Plot for Model 1")
plt.xlabel("Sample Index")
plt.ylabel("Values")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#model 2
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout

# Define the model
model2 = Sequential()

# Input Layer
model2.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))

# Hidden Layers
model2.add(Dense(32, activation='relu'))  # Hidden layer 1
# Dropout layer to prevent overfitting by randomly setting 20% of neurons to zero during training
model2.add(Dropout(0.2)) 
model2.add(Dense(16, activation='tanh'))  # Hidden layer 2

# Output Layer
model2.add(Dense(1, activation='linear'))  # Regression task

# Compile the model
model2.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.005),
              loss='mean_squared_error',
              metrics=['mean_absolute_error'])

# Print the model summary
model2.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
# EarlyStopping helps stop training a model early if the performance on the validation set stops improving

# - monitor='val_loss': monitors the validation loss during training
# - patience=5: allows training to continue for 5 more epochs after the last improvement in validation loss
# - restore_best_weights=True: restores the model weights from the epoch with the best validation loss
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# - epochs=500: maximum number of epochs for training (stops earlier if EarlyStopping is triggered)
# - batch_size=32: number of samples processed before updating the model weights
# - validation_split=0.2: reserves 20% of the training data for validation 
# - callbacks=[early_stopping]: ends training if validation loss stops improving
history = model2.fit(X_train, y_train, epochs=500, batch_size=32,
                    validation_split=0.2, callbacks=[early_stopping])
```

```python
import matplotlib.pyplot as plt

# Plot Loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss Over Epochs: Model 2')
plt.show()

# Plot MAE
plt.plot(history.history['mean_absolute_error'], label='Training MAE')
plt.plot(history.history['val_mean_absolute_error'], label='Validation MAE')
plt.xlabel('Epochs')
plt.ylabel('Mean Absolute Error: Model 2')
plt.legend()
plt.title('MAE Over Epochs')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Predict on the test set
y_pred = model2.predict(X_test)

# Scatter plot of true vs predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.6, edgecolor='k')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--')  # Line y=x
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('True vs Predicted Suicide Rates: Scatter Plot for Model 2')
plt.show()

# Line Plot
plt.figure(figsize=(12, 6))
plt.plot(range(len(y_test)), y_test, color='red', linestyle='-', label='True Values')
plt.plot(range(len(y_pred)), y_pred, color='blue', linestyle='-', label='Predicted Values')
plt.title("True vs Predicted Values: Line Plot for Model 2")
plt.xlabel("Sample Index")
plt.ylabel("Values")
plt.legend()
plt.grid(True)
plt.show()
```
